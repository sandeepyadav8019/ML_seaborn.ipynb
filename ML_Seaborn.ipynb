```python
# import thr necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# to ignore the warnings
from warnings import filterwarnings
```

```python
# load the dataset
df = sns.load_dataset('tips')

# the first five entries of the dataset
df.head()
```

```python
# load the dataset
import pandas as pd
df1=pd.read_csv(r"D:\BigData_All_content\csv\train_loan.csv")

# the first five entries of the dataset
df1.head()
```

```python
# set the background style of the plot
sns.set_style('whitegrid')
sns.distplot(df['total_bill'], kde = False, color ='red', bins = 30)
```

```python
sns.distplot(df1['ApplicantIncome'], kde = False, color ='red', bins = 30)
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df)
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df, kind ='kde')
# KDE shows the density where the points match up the most
```

```python
sns.jointplot(x ='ApplicantIncome', y ='LoanAmount', data = df1)
```

```python
#sns.jointplot(x ='total_bill', y ='tip', data = df, )
sns.jointplot(x ='ApplicantIncome', y ='LoanAmount', data = df1, kind ='kde')
```

```python
sns.pairplot(df1, hue ="Gender", palette ='coolwarm')
```

```python
sns.rugplot(df1['ApplicantIncome'])
```

```python
# plot the graph using the default estimator mean

sns.barplot(x ='Gender', y ='ApplicantIncome', data = df1, palette ='plasma')
```

```python
import numpy as np

# change the estimator from mean to standard deviation
sns.barplot(x ='Gender', y ='ApplicantIncome', data = df1, palette ='plasma', estimator = np.std)

```
